```python
# Import Libraries
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# read image
original_image = cv2.imread("man.png")
```

```python
# change the image into an 8-bit grayscale image
grayimg = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
```

```python
# height and width initialization
height = grayimg.shape[0] 
width = grayimg.shape[1] 
```

```python
# counter initialization
image_counter = np.zeros(256, dtype=int)
```

```python
# Looping to compute the histogram of the input image
for i in range(height): 
    for j in range(width): 
        image_counter[grayimg[i][j]] += 1 
```

```python
# Perform histogram equalization
equ_image = cv2.equalizeHist(grayimg)
```

```python
# equlization counter initialization
equ_counter = np.zeros(256, dtype=int) 
```

```python
# Looping to compute the histogram of the equalized image
for i in range(height): 
    for j in range(width): 
        equ_counter[equ_image[i][j]] += 1 
```

```python
# Result initialization
result = np.hstack([grayimg, equ_image]) 
```

```python
# DISPLAY (AND PRINT) THE ORIGINAL AND EQUALIZED IMAGES AS WELL ASTHEIR CORRESPONDING HISTOGRAMS, ALL IN ONE FIGURE.

plt.subplot(2, 1, 1)
plt.plot(image_counter, 'r', label="Histogram") 
plt.title("Low Contrast Image Histogram(Top) and Equalized Image Histogram(Down)")
plt.xlabel("Intensity") 
plt.ylabel("Counter") 
plt.axis([0, 256, 0, image_counter.max()]) 

plt.subplot(2, 1, 2)
plt.plot(equ_counter, 'r', label="Histogram") 
plt.xlabel("Intensity") 
plt.ylabel("Counter") 
plt.axis([0, 256, 0, equ_counter.max()]) 
plt.show()

plt.imshow(result, cmap='Greys_r') 
plt.title("Before and After Equalization")
plt.show()
```
